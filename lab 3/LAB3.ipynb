```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
import random
```

```python
image = data.camera()
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j] += random.randint(-20,20)
```

```python
plt.imshow(image, cmap = plt.cm.gray)
```

```python
mm = np.median(image)
print(mm)
```

```python
binary = image > mm
```

```python
plt.imshow(binary, cmap = plt.cm.gray)
```

```python
from skimage import feature

edges1 = feature.canny(image,sigma=3)
```

```python
plt.imshow(edges1, cmap = plt.cm.gray)
```

```python
def custom_canny(image):
  v = np.median(image)
  lower = int(max(0, 0.3 * v))
  upper = int(min(255, 1.3 * v))
  sigma = 1.5

  image_result = feature.canny(image, sigma, lower, upper)
	
  return image_result
```

```python
result = custom_canny(image)
plt.imshow(result, cmap = plt.cm.gray)
```
